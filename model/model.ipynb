```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

#Our sentences we like to encode
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.',
    'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```

```python
from tqdm.notebook import tqdm

def find_embeddings(df):
  embeddings = dict()
  for relation in tqdm(df['relation'].unique(), desc = "Relations"):
    buffer = []
    filtered_df = df.loc[df['relation'] == relation]
    for _, row in tqdm(filtered_df.iterrows(), desc = "Embedding"):
      sentence = row['arg1'] + ', ' + row['arg2']
      embedding = model.encode(sentence)
      buffer.append(embedding)
    
    mean_embedding = sum(buffer) / len(buffer)
    embeddings[relation] = mean_embedding
  return embeddings
```

```python
nerel_embeddings = find_embeddings(nerel_df)
```

```python
wikidata_embeddings = find_embeddings(wikidata_df)
```
